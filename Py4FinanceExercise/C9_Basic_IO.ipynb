```python
dtimes = np.arange('2019-01-01 10:00:00', '2025-12-31 22:00:00',
                    dtype='datetime64[m]') # creates an ndarray object with datetime as dtype
```

```python
len(dtimes)
```

```python
dty = np.dtype([('Date', 'datetime64[m]'),
                ('No1', 'f'),
                ('No2', 'f')]) # defines the special dtype object for the structured array
```

```python
data = np.zeros(len(dtimes), dtype=dty) # Instantiates an ndarray object with the special dtype
```

```python
data['Date'] = dtimes # populates the Date column
```

```python
a = np.random.standard_normal((len(dtimes),2)).round(4) # the dummy data set...
```

```python
data['No1'] = a[:,0] # ...which populate the No1
data['No2'] = a[:,1] # and No2 columns
```

```python
data.nbytes # the size of the structured array in bytes
```

```python
%%time
temp = q('SELECT * FROM numbers').fetchall() # retrieves all the rows from the table in a single step
print(temp[:3])
```

```python
%%time
query = 'SELECT * FROM numbers WHERE No1 > 0 AND No2 < 0'
res = np.array(q(query).fetchall()).round(3) # retrieves a selection of the rows and transforms it to an ndarray object
```

```python
res = res[::100]
plt.figure(figsize=(10, 6))
plt.plot(res[:, 0], res[:,1], 'ro') #plots a subset of the query result
```
